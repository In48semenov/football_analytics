```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
result_folder = './Test_results/'
```

```python
df_write = pd.read_csv(os.path.join(result_folder, 'S3_wrtie.csv'))
df_read = pd.read_csv(os.path.join(result_folder, 'S3_read.csv'))
df_write.dropna(how='any', inplace=True)
df_read.dropna(how='any', inplace=True)
```

```python
df_write
```

```python
df_read
```

```python
df_write_mean = df_write[['size_file', 'time', 'IO']].groupby(['size_file']).mean()
df_read_mean = df_read[['size_file', 'time', 'IO']].groupby(['size_file']).mean()
```

```python
df_write_mean
```

```python
df_read_mean
```

```python
x = np.arange(len(df_write_mean))  
width = 0.35 

fig, ax = plt.subplots(figsize=(12,8))
rects1 = ax.bar(x - width/2, df_write_mean['time'], width, label='write')
rects2 = ax.bar(x + width/2, df_read_mean['time'], width, label='read')

ax.set_xlabel(' , ')
ax.set_ylabel('  , ')
plt.xticks(
    [idx for idx in range(len(df_write_mean))],
    [f'{round(size, 2)} ' for size in df_write_mean.index])
ax.legend()
plt.title('     ', fontsize=15)
plt.grid(True);
plt.savefig('     ')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(np.arange(len(df_write_mean)), 100 - df_read_mean['time']/df_write_mean['time']*100)
ax.set_xlabel(' , ')
ax.set_ylabel('  %')
plt.xticks(
    [idx for idx in range(len(df_write_mean))],
    [f'{round(size, 2)} ' for size in df_write_mean.index])
plt.title('       %', fontsize=15)
plt.grid(True);
plt.savefig('       %')
```

```python
x = np.arange(len(df_write_mean))  
width = 0.35 

fig, ax = plt.subplots(figsize=(12,8))
rects1 = ax.bar(x - width/2, df_write_mean['IO'], width, label='write')
rects2 = ax.bar(x + width/2, df_read_mean['IO'], width, label='read')

ax.set_xlabel(' , ')
ax.set_ylabel('  , /')
plt.xticks([round(size, 2)*1e-3 for size in df_write['size_file']])
plt.xticks(
    [idx for idx in range(len(df_write_mean))],
    [f'{round(size, 2)} MB' for size in df_read_mean.index])
ax.legend()
plt.title('  ', fontsize=15)
plt.grid(True);
plt.savefig('  ')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.bar(np.arange(len(df_write_mean)), 100 - df_write_mean['IO']/df_read_mean['IO']*100)
ax.set_xlabel(' , ')
ax.set_ylabel('  %')
plt.xticks(
    [idx for idx in range(len(df_write_mean))],
    [f'{round(size, 2)} ' for size in df_write_mean.index])
plt.title('        %', fontsize=15)
plt.grid(True);
plt.savefig('        %')
```
